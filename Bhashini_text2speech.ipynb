```python
# text to speech , play audio function
def tts_bhashini(HEAD, IP_TEXT):

    import io
    from IPython.display import Audio, display

    URL = "https://tts.bhashini.ai/v1/synthesize"
    data = {
        "text": IP_TEXT,
        "languageId": "en",
        "voiceId": 0
    }

    response = requests.post(URL, json=data, headers= HEAD)
    response.raise_for_status()

    audio_stream = io.BytesIO(response.content)
    display(Audio(audio_stream.read(), rate=44100, autoplay=True))

    # Make the audio file from binary data
    # audio_file = 'Audio.mp3'
    # with open(audio_file , '+wb') as file:
    #     file.write(response.content)
    # return audio_file

# def play_audio(file_path):
#     from IPython.display import Audio, display
#     display(Audio(file_path, autoplay=True))

# translation function
def translation_bhashini(HEAD, IP_TEXT):

    URL = "https://tts.bhashini.ai/v1/translate"
    data = {
    "inputText": IP_TEXT ,
    "inputLanguage": "English",
    "outputLanguage": "Hindi"
    }

    response = requests.post(URL, json=data, headers= HEAD)
    response.raise_for_status()
    return response.text

# transliteration function
def transliteration_bhashini(HEAD, IP_TEXT):
    URL = "https://tts.bhashini.ai/v1/transliterate"
    data = {
        "inputText": IP_TEXT,
        "inputScriptId": "Taml",
        "outputScriptId": "Knda"
    }

    response = requests.post(URL, json=data, headers= HEAD)
    response.raise_for_status()
    return response
```
